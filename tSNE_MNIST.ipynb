```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sn
from sklearn.manifold import TSNE

import io
d0 = pd.read_csv(io.StringIO(uploaded['train_MNIST.csv'].decode('utf-8')))
```

```python
l = d0['label']
d = d0.drop("label",axis=1)
from sklearn.preprocessing import StandardScaler
standardized_data = StandardScaler().fit_transform(d)
```

# t-SNE with 15k points

```python
data_15k = standardized_data[0:15000,:]
labels_15k = l[0:15000]
```

```python
# perplexity=2, n_iter=1000
model = TSNE(n_components=2, perplexity=2, n_iter=1000, random_state=0)
tsne_data = model.fit_transform(data_15k)
tsne_data = np.vstack((tsne_data.T, labels_15k)).T
tsne_df = pd.DataFrame(data=tsne_data, columns=("Dim_1", "Dim_2", "label"))
sn.FacetGrid(tsne_df, hue="label", size=6).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
plt.show()
```
